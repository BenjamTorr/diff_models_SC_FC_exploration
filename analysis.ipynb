```python
from SCFC_data_extraction import *
import torch
import pandas as pd
import plotly.express as px
from corticalSCFCDataset import *
from diffusionModel import *
from networkModel import *
from utils import *
import model_eval
import random
import numpy as np

SEED = 0
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
```

```python
ids_FC, cortical_FC, cortical_SC = load_cortical_SC_FC(transform_SC = False, transform_FC = False)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}\t" + (f"{torch.cuda.get_device_name(0)}" if torch.cuda.is_available() else "CPU"))
```

```python

batch_size = 128
n_epochs = 100
lr = 0.001
store_path = "test.pt"
loader = DataLoader(corticalSCFCDataset(SC_matrices= cortical_SC, FC_matrices= cortical_FC), batch_size, shuffle=True)


n_steps, min_beta, max_beta = 1000, 10 ** -4, 0.02  # Originally used by the authors
ddpm = MyDDPM(MyUNet(n_steps, feat_dim = 1), n_steps=n_steps, min_beta=min_beta, max_beta=max_beta, device=device)
```

```python
optim = torch.optim.Adam(ddpm.parameters(), lr=lr)
import time
start_time = time.time()

n_epochs=100

training_loop(ddpm, loader, n_epochs, optim, device, display=False, store_path="models/" + store_path)

print('Execution time:', '{:5.2f}'.format(time.time() - start_time), 'seconds')
```

# After Training

```python
best_model = MyDDPM(MyUNet(), n_steps=n_steps, device=device)
best_model.load_state_dict(torch.load("models/" + 'ddpm_mnist_raw.pt', map_location=device))
best_model.eval()
print("Model loaded")
```
